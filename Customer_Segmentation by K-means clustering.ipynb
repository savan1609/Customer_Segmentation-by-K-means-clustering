### Importing Libraries

```python
import  numpy as np
import  pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
customer = pd.read_csv('marketing_campaign.csv')
```

```python
customer.head()
```

```python
customer.describe()
```

```python
customer.info()
```

```python
customer = customer.dropna()
```

```python
customer["Dt_Customer"] = pd.to_datetime(customer["Dt_Customer"])
```

```python
customer.head()
```

```python
dates = []
for i in customer["Dt_Customer"]:
    i = i.date()
    dates.append(i)
```

```python
customer.head()
```

### Feature Engineering

```python
#Created a feature "Customer_For"
days = []
d1 = max(dates) #taking it to be the newest customer
for i in dates:
    delta = d1 - i
    days.append(delta)
customer["Customer_For"] = days
customer["Customer_For"] = pd.to_numeric(customer["Customer_For"], errors="coerce")
```

```python
customer["Marital_Status"].value_counts()
```

```python
customer["Education"].value_counts()
```

```python
#Age of customer today
customer["Age"] = 2021-customer["Year_Birth"]
```

```python
#Total spendings on various items
customer["Spent"] = customer["MntWines"]+ customer["MntFruits"]+ customer["MntMeatProducts"]+ customer["MntFishProducts"]+ customer["MntSweetProducts"]+ customer["MntGoldProds"]
```

```python
#Deriving living situation by marital status"Alone"
customer["Living_With"] = customer["Marital_Status"].replace({"Married":"Partner", "Together":"Partner", "Absurd":"Alone", "Widow":"Alone", "YOLO":"Alone", "Divorced":"Alone", "Single":"Alone",})
```

```python
#Total children living in the household
customer["Children"] = customer["Kidhome"]+customer["Teenhome"]
```

```python
#Total members in the householde
customer["Family_Size"] = customer["Living_With"].replace({"Alone": 1, "Partner":2})+ customer["Children"]
```

```python
#Feature pertaining parenthood
customer["Is_Parent"] = np.where(customer.Children> 0, 1, 0)
```

```python
#Segmenting education levels in three groups
customer["Education"] = customer["Education"].replace({"Basic":"Undergraduate","2n Cycle":"Undergraduate", "Graduation":"Graduate", "Master":"Postgraduate", "PhD":"Postgraduate"})
```

```python
#For clarity
customer = customer.rename(columns={"MntWines": "Wines","MntFruits":"Fruits","MntMeatProducts":"Meat","MntFishProducts":"Fish","MntSweetProducts":"Sweets","MntGoldProds":"Gold"})
```

```python
#Dropping some of the redundant features
to_drop = ["Marital_Status", "Dt_Customer", "Z_CostContact", "Z_Revenue", "Year_Birth", "ID"]
customer = customer.drop(to_drop, axis=1)
```

```python
new = customer.copy()
# creating a subset of dataframe by dropping the features on deals accepted and promotions
cols_del = ['AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1','AcceptedCmp2', 'Complain', 'Response']
new = new.drop(cols_del, axis=1)
```

```python
new
```

```python
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
SS.fit(new)
scaled_new = pd.DataFrame(SS.transform(new),columns= new.columns)
```

```python
scaled_new
```

### PCA

```python
#Use PCA to reduce dimentions features to 3
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
pca.fit(scaled_new)
PCA_new = pd.DataFrame(pca.transform(scaled_new), columns=(["col1","col2", "col3"]))
PCA_new.describe().T
```

```python
#A 3D Projection Of Data In The Reduced Dimension
x =PCA_new["col1"]
y =PCA_new["col2"]
z =PCA_new["col3"]
fig = plt.figure(figsize=(20,16))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(x,y,z, marker="o" )
```

### Clustering

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state = 42)
    kmeans.fit(PCA_new)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('wcss')
```

```python
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(PCA_new)
print(y_kmeans)
PCA_new["Clusters"] = y_kmeans
customer["Clusters"] = y_kmeans
```

```python
#Plotting the clusters
fig = plt.figure(figsize=(20,16))
ax = plt.subplot(111, projection='3d', label="bla")
ax.scatter(x, y, z, s=40, c=PCA_new["Clusters"], marker='o')
ax.set_title("The Plot Of The Clusters")
```

```python
pl = sns.countplot(x=customer["Clusters"])
pl.set_title("Distribution Of The Clusters")
```

```python
pl = sns.scatterplot(x=customer["Spent"], y=customer["Income"], hue=customer["Clusters"])
pl.set_title("Cluster's Profile Based On Income And Spending")
plt.legend()
```
